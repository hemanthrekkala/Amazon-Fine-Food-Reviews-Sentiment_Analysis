```python
ml['ProductId'] = ml['ProductId'].apply(lambda x:x.replace('ProductId:',""))
ml['ProductId'] = ml['ProductId'].apply(lambda x:x.replace('\n',""))
ml['ProductId'] = ml['ProductId'].apply(lambda x:x.replace(':',""))
ml['ProfileName'] = ml['ProfileName'].apply(lambda x:x.replace(':',""))
ml['HelpfulnessNumerator'] = ml['HelpfulnessNumerator'].apply(lambda x:x.replace('HelpfulnessNumerator:',""))
ml['HelpfulnessDenominator'] = ml['HelpfulnessDenominator'].apply(lambda x:x.replace('HelpfulnessDenominator:',""))
ml['Score'] = ml['Score'].apply(lambda x:x.replace('Score:',""))
ml['Time'] = ml['Time'].apply(lambda x:x.replace('Time:',""))
ml['ReviewSummary'] = ml['ReviewSummary'].apply(lambda x:x.replace('ReviewSummary:',""))
ml['ReviewText'] = ml['ReviewText'].apply(lambda x:x.replace('ReviewText:',""))
ml['ProfileName'] = ml['ProfileName'].apply(lambda x:x.replace('\n',""))
ml['HelpfulnessNumerator'] = ml['HelpfulnessNumerator'].apply(lambda x:x.replace('\n',""))
ml['HelpfulnessDenominator'] = ml['HelpfulnessDenominator'].apply(lambda x:x.replace('\n',""))
ml['Score'] = ml['Score'].apply(lambda x:x.replace('\n',""))
ml['Time'] = ml['Time'].apply(lambda x:x.replace('\n',""))
ml['ReviewSummary'] = ml['ReviewSummary'].apply(lambda x:x.replace('\n',""))
ml['UserId'] = ml['UserId'].apply(lambda x:x.replace('\n',""))
ml['UserId'] = ml['UserId'].apply(lambda x:x.replace('UserId:',""))
ml['ProfileName'] = ml['ProfileName'].apply(lambda x:x.replace('ProfileName',""))
ml['UserId'] = ml['UserId'].apply(lambda x:x.replace(':',""))
```

```python
ml['HelpfulnessNumerator'] = ml['HelpfulnessNumerator'].astype(str).astype('int64')
```

```python
ml['HelpfulnessDenominator'] = ml['HelpfulnessDenominator'].astype(str).astype('int64')
```

```python
ml['Score'] = ml['Score'].astype(str).astype('int64')
```

```python
amzon_review = ml.drop_duplicates()
```

```python
amzon_review.info()
```

```python
ml['usefulness'] = ml['HelpfulnessNumerator']/ml['HelpfulnessDenominator']
```

```python
usefulness = []
for i in ml['usefulness']:
    if i > 0.75:
        usefulness.append(">75%")
    elif i < 0.25:
        usefulness.append("<25%")
    elif i >= 0.25 and i <= 0.75:
        usefulness.append("25-75%")
    else:
        usefulness.append("useless")
ml['usefulness']  = usefulness   
```

```python
word_count = []
for i in ml['ReviewText']:
    word_count.append(len(i.split()))
ml['word_count'] = word_count    
```

```python
sentiment = []
for i in amzon_review['Score']:
    if i > 3:
        sentiment.append('positive')
    elif i < 3:
        sentiment.append('negative')
    else:
        sentiment.append('not')
amzon_review['sentiment'] = sentiment        
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
comment_words = ''
stopwords = set(STOPWORDS)

for val in pos['ReviewSummary']:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    comment_words += " ".join(tokens)+" "
    
wordcloud = WordCloud(width = 600, height = 600,
                background_color ='red',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)    
    
plt.figure(figsize = (10, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()        
```
